```python
import matplotlib.pyplot as plt

# Sample data: Population by gender for a specific year
years = ['2000', '2005', '2010', '2015', '2020']
male_population = [500000, 520000, 540000, 560000, 580000]
female_population = [480000, 500000, 520000, 540000, 560000]

# Bar width
bar_width = 0.35

# Bar positions
r1 = range(len(years))
r2 = [x + bar_width for x in r1]

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(r1, male_population, color='red', width=bar_width, edgecolor='grey', label='Male')
plt.bar(r2, female_population, color='pink', width=bar_width, edgecolor='grey', label='Female')

# Adding titles and labels
plt.title('Population Distribution by Gender Over Years')
plt.xlabel('Year')
plt.ylabel('Population')
plt.xticks([r + bar_width / 2 for r in range(len(years))], years)
plt.legend()

# Show the plot
plt.show()
```
